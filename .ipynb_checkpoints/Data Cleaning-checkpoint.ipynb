```python
#Dependencies
import pandas as pd
import numpy as np
import datetime
```

```python
#Filepath
file = "Resources/Cleaned Data Date.csv"
```

```python
#Read in CSV File
datesRawData_DF = pd.read_csv(file, dtype = str)
datesRawData_DF.head()
```

```python
#Drop NAN
datesNANData_DF = datesRawData_DF.dropna()
datesNANData_DF.head()
```

```python
#Remove unneeded columns
datesCondensed_DF = datesNANData_DF.reindex(columns = ["WINTER", "Winter Start", "CLOSED", "Closed Full Date", "OPENED", "Opened Full Date", "DAYS"])
datesCondensed_DF = datesCondensed_DF.reset_index(drop=True)
datesCondensed_DF.head()
```

```python
#Julian Date Functions
def JdateC (row):
    month, day, year = (int(x) for x in row['Closed Full Date'].split('/'))
    ans = datetime.date(year, month, day)
    z = ans.strftime("%j")
    return(z)

def JdateO (row):
    month, day, year = (int(x) for x in row['Opened Full Date'].split('/'))
    ans = datetime.date(year, month, day)
    z = ans.strftime("%j")
    return(z)
```

```python
#Run JdateC
datesCondensed_DF['Julian Closed'] = datesCondensed_DF.apply(JdateC, axis = 1)
```

```python
#Run JdateO
datesCondensed_DF['Julian Opened'] = datesCondensed_DF.apply(JdateO, axis = 1)
```

```python
#Test Julian Dates
datesCondensed_DF.head()
```

```python
#Organize columns
datesClean_DF = datesCondensed_DF.reindex(columns = ["WINTER", "Winter Start", "CLOSED", "Closed Full Date", "Julian Closed", "OPENED", "Opened Full Date", "Julian Opened", "DAYS"])
datesClean_DF = datesClean_DF.reset_index(drop=True)
datesClean_DF.head()
```
